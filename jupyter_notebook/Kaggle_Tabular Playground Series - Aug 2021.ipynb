## 外れ値の処理

```python
#外れ値（平均値から標準偏差3倍分離れた値）を3σの値に変換する関数を定義
def del_idx_extract(df):
    
    for i in range(len(df.columns)):
        col_std = df.iloc[:, i].std()
        col_mean = df.iloc[:, i].mean()
        outliers_max = col_mean + (col_std * 3)#標準偏差から3倍分加えた値を最大値として定義
        outliers_min = col_mean - (col_std * 3)#標準偏差から3倍分引いた値を最小値として定義
        
        for j in range(len(df)):
            if df.iat[j, i] > outliers_max:
                df.iat[j, i] = outliers_max #平均値から3σ以上大きい値をoutliers_maxに変換
            elif df.iat[j, i] < outliers_min:
                df.iat[j, i] = outliers_min #平均値から3σ以上小さい値をoutliers_minに変換
```

```python
del_idx_extract(df_zs_X)
```
